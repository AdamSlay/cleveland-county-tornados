```python
tornados = geopandas.read_file(
    '../data/1950-2021-torn-initpoint'
)
```

```python
ok = ok.to_crs("EPSG:3395")
```

```python
ok.plot()
```

```python
tornados = tornados.to_crs("EPSG:3395")
```

```python
ok_tors = tornados[(tornados['st'] == 'OK') & (tornados['yr'] == 2021)]
ok_tors
```

```python
fig = plt.figure(1, figsize=(12,6))
ax = fig.add_subplot()
fig.suptitle('OK Tornados (bogus)', fontsize=16)

ok.plot(ax=ax, color='black', linewidth=.8)
ok_tors.plot(ax=ax, color='red', marker='v', markersize=20)
```

```python
county = ok[ok['COUNTYFP'] == '027']
```

```python
fig = plt.figure(1, figsize=(12,6))
ax = fig.add_subplot()
fig.suptitle('OK Tornados (bogus)', fontsize=16)

ok.plot(ax=ax, color='black', linewidth=.8)
county.plot(ax=ax, color='green')
ok_tors.plot(ax=ax, color='red', marker='v', markersize=20)
```
